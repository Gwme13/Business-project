```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import re
import contractions
from fpdf import FPDF
from PIL import Image
import nltk
import google.generativeai as genai


# Define the path to the McDonald's reviews dataset
macdonalds_path = os.path.join('dataset', 'McDonald_s_Reviews.csv')

# Read the CSV file with a specified encoding
macdonalds_df = pd.read_csv(macdonalds_path, encoding='latin1')

# Display the first few rows to understand the structure
print(macdonalds_df.head())

# Function to call the Google AI API
def generate_text(company_name, file_path, **kwargs):
    prompt = f""" 
    You are a virtual assistant to the CEO of {company_name}. I will provide you with reviews in text format. Analyze them and provide:
        1) Main Topics.
        2) Assign a rank of positive or negative to each.
        3) For the formulation of a new strategy in business, identify my company's main problems (worst topic) and propose technical-economic feedback.
    """
    
    # Add the additional information to the prompt if available
    if len(kwargs) > 0:
        for arg in kwargs:
            if isinstance(kwargs[arg], str):
                prompt += f"\n{kwargs[arg]}"
    
    # API Key configuration (replace with actual key)
    API_KEY = "AIzaSyBzjTSU97Yedj0yo5GDLxuUQVxxCWDunVk"
    genai.configure(api_key=API_KEY)
    
    model = genai.GenerativeModel('gemini-1.5-flash')
    
    sample_file = genai.upload_file(path=file_path, display_name=file_path)
    print(f"Uploaded file '{sample_file.display_name}' as: {sample_file.uri}")
    
    # Generate content using the uploaded document and the prompt
    response = model.generate_content([sample_file, prompt])
    
    # Print or return the generated content
    print(response.text)
    return response.text

# Generic function to process Reviews and Ratings
def process_reviews_and_ratings(df, review_col, rating_col, output_pdf, api_function=None, company_name=None):
    # 1. Preprocessing: Clean reviews
    df[review_col] = df[review_col].str.lower()  # Convert reviews to lowercase
    df[review_col] = df[review_col].apply(lambda x: contractions.fix(x))  # Correct contractions
    df[review_col] = df[review_col].apply(lambda x: re.sub(r'[^\w\s]', '', x))  # Remove punctuation
    
    # Optional: Tokenize reviews
    df['tokenized_review'] = df[review_col].apply(nltk.word_tokenize)
    # 2. Transform the ratings column to integers
    # Clean the rating column by extracting the numeric part
    df[rating_col] = df[rating_col].apply(lambda x: re.findall(r'\d+', str(x))[0])  # Extract numeric part
    df[rating_col] = df[rating_col].astype(int)  # Convert to integer

    # 2. Transform the ratings column to integers
    df[rating_col] = df[rating_col].apply(lambda x: int(x))

    # 3. Pie chart of rating distribution
    plt.figure(figsize=(8, 8))
    df[rating_col].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("muted"))
    plt.title(f'{rating_col} Distribution')
    plt.ylabel('')
    plt.savefig('rating_distribution.png')
    plt.show()

    # 4. Generate WordCloud
    def plot_wordcloud(text):
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.savefig('wordcloud.png')
        plt.show()

    # Concatenate all reviews for word cloud generation
    plot_wordcloud(' '.join(df['tokenized_review'].apply(lambda tokens: ' '.join(tokens))))

    # 5. Save reviews to a text file
    reviews_file = os.path.join('.', 'reviews.txt')
    with open(reviews_file, 'w') as file:
        for review in df[review_col]:
            file.write(review + '\n')

    # 6. Call the external API (Optional)
    if api_function and company_name:
        response_text = api_function(company_name, reviews_file)
        # You can include this response text in the PDF if needed

    # 7. Create PDF with text and images
    create_pdf(reviews_file, 'wordcloud.png', output_pdf)

# Create PDF function as previously defined
def create_pdf(txt_file, img_file, output_pdf):
    # Create PDF object
    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)
    
    # Add a page for the text
    pdf.add_page()
    pdf.set_font("Arial", size=12)

    # Write the text content into PDF
    with open(txt_file, 'r', encoding='utf-8') as f:
        for line in f:
            pdf.multi_cell(0, 10, line)

    # Add a new page for the word cloud image
    pdf.add_page()

    # Load and resize image
    image = Image.open(img_file)
    image_width, image_height = image.size
    max_width, max_height = 150, 150  # Maximum size for PDF

    # Calculate resize ratio
    width_ratio = max_width / image_width
    height_ratio = max_height / image_height
    scale_ratio = min(width_ratio, height_ratio)

    new_width = int(image_width * scale_ratio)
    new_height = int(image_height * scale_ratio)

    # Add the image to the PDF
    pdf.image(img_file, x=10, y=None, w=new_width, h=new_height)

    # Save the final PDF
    pdf.output(output_pdf)

# Example usage with McDonald's dataset
process_reviews_and_ratings(macdonalds_df, 'review', 'rating', 'output_mcdonalds.pdf', generate_text, company_name="McDonald's")
```
